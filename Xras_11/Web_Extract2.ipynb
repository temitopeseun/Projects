```python
def print_date(start_year):
    from datetime import date, timedelta, datetime
    import pandas as pd

    start_date = datetime(int(start_year), 1, 1)
    curr_date = start_date
    while curr_date.year == start_date.year:
        yield [curr_date.strftime("%Y"), curr_date.strftime("%m"), curr_date.strftime("%d")] 
        curr_date += timedelta(days=1) #add one day to the date

```
