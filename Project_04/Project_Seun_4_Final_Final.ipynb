```python
import requests
import pandas as pd
import requests
from bs4 import BeautifulSoup
from bs4 import BeautifulSoup as soup
from bs4 import SoupStrainer
import re
import os
```

```python

main_URL = 'https://www.lyrics.com'


# Artist_1: Jay-Z
url_1 = 'https://www.lyrics.com/artist/Jay-Z/'

# Artist_2: Tupac
url_2 = 'https://www.lyrics.com/artist/Tupac-Shakur/'
```

```python
# connect to the song server and extract data

response_1 = requests.get(url_1)
JayZ_page = soup(response_1.content, "html.parser")


response_2= requests.get(url_2)
Tupac_page = soup(response_2.content, "html.parser")
```

```python
#extracting the name of the artist

artist1 = JayZ_page.h1.a.text
artist2 = Tupac_page.h1.a.text
```

```python
#  Get JayZ and Tupac lyric links class

links1 = JayZ_page.find_all(class_="tal qx")
#links1

links2 = Tupac_page.find_all(class_="tal qx")
```

```python
# loop every link to extract the each song lyrics urls for Jay-Z

main_URL = 'https://www.lyrics.com'

lyrik1=[]

for i in links1:
    
    lyric_URL_link = i.a['href']
    
    song_URL = main_URL + lyric_URL_link
   
    lyrik1.append(song_URL)
    
    print(song_URL)
```

```python
# loop every link to extract the each song lyrics urls for Tupac

lyrik2=[]

for j in links2:
    
    lyric_URL_link2 = j.a['href']
    
    song_URL2 = main_URL + lyric_URL_link2
    
     
    lyrik2.append(song_URL2)
    
    print(song_URL2)
```

```python
# Define links as variables

L1=lyrik1
L2=lyrik2

```

```python
# This code remove double or duplicate lyrics in JayZ

my_set = set()
my_list= []
for x in L1:
    y = x.split('/')[-1]
    if y not in my_set:
        my_list.append(x)
        my_set.add(y)
```

```python
# This code remove double or duplicate lyrics in Tupac

my_set2 = set()
my_list2 = []
for xi in L2:
    yi = xi.split('/')[-1]
    if yi not in my_set2:
        my_list2.append(xi)
        my_set2.add(yi)
        
# len(my_list2) # for the length
```

```python
#extracting certian number of lyrics from JayZ and Tupac songs

my_list=my_list[:50]
my_list2=my_list2[:50]
```

```python

#extracting lyrics title from url of JayZ songs, and removing special Characters

song_title1=[]

for link in my_list:
    response = requests.get(link, headers=headers).text
    songs_soup = BeautifulSoup(response, 'html.parser')
    lyrics = songs_soup.find(class_='lyric-body').text
    song_title = link.split('/')[6].replace("+", " ")
    song_title = song_title.split('/')[0].replace("%27", "'")
    song_title = song_title.split('/')[0].replace("%5B%2A%5D", " ")
    song_title = song_title.split('/')[0].replace("%3", " ")
   
    song_title1.append(song_title)

    with open(f'{song_title}.txt', 'w', encoding='utf-8') as f:
        f.write(lyrics)
       
    #print(song_title)
```

```python
#extracting lyrics title from url of Tupac songs and removing special characters

song_title2=[]

for link in my_list2:
    
    response = requests.get(link, headers=headers).text
    songs_soup = BeautifulSoup(response, 'html.parser')
    lyrics = songs_soup.find(class_='lyric-body').text
    song_title = link.split('/')[6].replace("+", " ")
    song_title = song_title.split('/')[0].replace("%27", "'")
    song_title = song_title.split('/')[0].replace("%C3%AD", " ")
    song_title = song_title.split('/')[0].replace("%26", "&")
    song_title = song_title.split('/')[0].replace("%28", "(")
    song_title = song_title.split('/')[0].replace("%29", ")")
    song_title = song_title.split('/')[0].replace("%5B", "(")
    song_title = song_title.split('/')[0].replace("%5D", ")")
    song_title = song_title.split('/')[0].replace("%2C", "")
   
    song_title2.append(song_title)
    
    with open(f'{song_title}.txt', 'w', encoding='utf-8') as f:
        f.write(lyrics)
        
        #print(song_title)
        
       
```

```python
# Functions for extracting lyrics

def lyrics_Links_Extract(MYLIST):
    
    lyrics =[]
    
    for link in MYLIST:
        l = requests.get(link)
        url = soup(l.content, parse_only=SoupStrainer(class_='lyric-body'))
        lyrics.append(url.text)
    
    return lyrics
```

```python
#extracting JayZ lyrics 

lyrics1=lyrics_Links_Extract(my_list)
```

```python
#extracting Tupac lyrics 

lyrics2=lyrics_Links_Extract(my_list2)
```

```python
# Make a DataFrame df_JayZ and df_Tupac

df_JayZ = pd.DataFrame({'artist': artist1, 'songs': song_title1, 'url': my_list,'lyrics': lyrics1})

df_Tupac = pd.DataFrame({'artist': artist2, 'songs': song_title2, 'url': my_list2,'lyrics': lyrics2})

df = pd.concat([df_JayZ, df_Tupac], ignore_index=True)

```

# Tokenization and Lemmatization

* Tokenization is the process of splitting a text block or tokens.
* The text block or tokens are in smaller text chunks of words or phraces that is meaninngful. 
* Lemmatization is the process of reverting a word back to it's root/base word in the dictionary. 
* Lemmatization example: kisses ==> kiss, feet ==> foot, wolves ==> wolf

```python
# Combining the 2 singers url

SINGERS_url=my_list + my_list2
#len(SINGERS_url)
```

```python
import requests

Coupus=[]

for songlinx in SINGERS_url:
    
    URLS= songlinx
    SongResponse= requests.get(URLS)
 
    Singers_LYRICS = BeautifulSoup(SongResponse.text)
    
    for title in Singers_LYRICS.find_all(class_='lyric-body'):
    
     Coupus.append(title.text)
```

```python
Coupus=[s.lower() for s in Coupus]
Coupus
#len(Coupus)
```

```python
#tokenize and lemmaztize (i.e.get lyrics cleaned up)

from nltk.stem import WordNetLemmatizer

from nltk.tokenize import TreebankWordTokenizer ## very good tokenizer for english, considers sentence structure


CLEAN_CORPUS = []

tokenizer = TreebankWordTokenizer()
lemmatizer = WordNetLemmatizer()


for doc in Coupus:
    tokens = tokenizer.tokenize(text=doc)
    clean_doc = " ".join(lemmatizer.lemmatize(token) for token in tokens)
    CLEAN_CORPUS.append(clean_doc)
```

```python
CLEAN_CORPUS
```

# Using a machine learning classification algorithm

 * Text Vectorization:Transform the text-data into machine readable (numerics)

 *  Text Vectorization approach:TF-IDF

 *  TF - Term Frequency (% count of a word 𝑤 in doc 𝑑 ), IDF - Inverse Document Frequency
 
 *  We use a TfidfVectorizer function which is a single step.


# Vectorization

* Words that do not provide any useful information to decide in which category a text 
  should be classified are refered to as Stopwords.

* These words would be removed from the data in order to reduce the length of the vectors

```python
import nltk
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from nltk.corpus import stopwords
nltk.download('stopwords', quiet=True)
```

```python
STOPWORDS = stopwords.words('english')
vectorizer = CountVectorizer(stop_words=STOPWORDS)
vectors = vectorizer.fit_transform(CLEAN_CORPUS)
vectors.todense()
```

```python
Vec_gfn=vectorizer.get_feature_names()
```

```python
# let's put the vectorized lyrics in a pandas dataframe with the labels

df_VECTOR = pd.DataFrame(vectors.todense(), columns=Vec_gfn, index=df.artist)
df_VECTOR
```

```python

def fit_model(X_train, y_train):
    """Return a trained model based on the corpus.
    Take the corpus and labels."""

    model = make_pipeline(
        TfidfVectorizer(),
        MultinomialNB()
    )

    fitted_model = model.fit(X_train, y_train)

    return fitted_model
```

```python
def make_predictions(model, X_test):
    """Returns the class with most probability and also correspondent probabilities.
    Takes a fitted model and X_test (unseen data)."""

    X_test = [X_test]
    
    predictions = model.predict(X_test)
    probs = model.predict_proba(X_test)
    

    #return predictions, probs
    return print('The song was probably written by:', predictions[0], 'with an accuracy of', (probs[0][0] * 100).round(1), '%.')
```

```python
Y_true = ["Jay-Z"]*len(df[df.artist == 'Jay-Z']) + ["Tupac Shakur"]*len(df[df.artist == 'Tupac Shakur'])
```

```python
fitted_model = fit_model(X_train = CLEAN_CORPUS, y_train = Y_true)
```

```python
X_test = "Bad Boy murdered"

predictions = make_predictions(model = fitted_model, X_test = X_test)

predictions
```
