### Task 1

From SPICED Data Science page, extract the text description for each of part of the curriculum.

#### Step 1: Download the text of the HTML from the website

```python
import requests
```

```python
url = 'https://www.spiced-academy.com/en/program/data-science'
```

```python
response = requests.get(url)
response
```

#### Step 2: Convert the raw HTML into a BeautifulSoup object, so that the data can be easily parsed

```python
from bs4 import BeautifulSoup
```

```python
spiced_soup = BeautifulSoup(response.text)
```

#### Step 3. Use the BeautifulSoup object to parse the HTML document tree down to the tag / element that contains the data you want

```python
type(spiced_soup)
```

```python
print(spiced_soup.prettify())
```

```python
def pp(soup):
    print(soup.prettify())
```

```python
pp(spiced_soup)
```

```python
spiced_soup.title.text
```
